```python
import numpy as np
import pandas as pd
import seaborn as sns
import warnings # current version of seaborn generates a bunch of warnings that we'll ignore
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
import keras
from keras.wrappers.scikit_learn import KerasClassifier #use Sequential Keras models as part of your Scikit-Learn workflow via the wrappers

sns.set(style="white", color_codes=True)
data = pd.read_csv(".\/Iris-ANN/Iris.csv")
```

```python
# Wrangling
# Encoding categorical values with LabelEncoder: replaced Iris-setosa with 0, Iris-versicolor 1 and Iris-virginica 2 data.set_value(data['Species']=='Iris-setosa',['Species'],0)
data.set_value(data['Species']=='Iris-setosa',['Species'],0)
data.set_value(data['Species']=='Iris-versicolor',['Species'],1)
data.set_value(data['Species']=='Iris-virginica',['Species'],2)
data = data.apply(pd.to_numeric)
```

```python
# Splitting the dataset into the Training set and Test set

train=data.sample(frac=0.8,random_state=200)
test=data.drop(train.index)

X_train = train[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
X_test = test[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y_train = train['Species']
y_test = test['Species']

X_train_array = X_train.as_matrix()
X_train = X_train_array
```

```python
# Encoding categorical variables
from keras.utils import np_utils
y_train = np_utils.to_categorical(y_train)
```

```python
# Build the model in keras
# Keras: neural network python library running with Tensorflow (here) as backend

# The Sequential model is a linear stack of layers.
from keras.models import Sequential
# Dense will add hidden layers 
from keras.layers import Dense, Activation

# Initializing Neural Network
model = Sequential()
# Adding the input layer and the first hidden layer
model.add(Dense(output_dim=10, input_dim=4))
model.add(Activation("relu"))
# Adding the second hidden layer
model.add(Dense(output_dim=10))
model.add(Activation("relu"))
# Adding the output layer
model.add(Dense(output_dim=3))
model.add(Activation("softmax"))
```

```python
# Configure the learning proces with the optimizer algorithm (find optimal weights), the loss function and accuracy metrics
model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

```python
# Train the model
# Batch size: number of observation after which you want to update weight
# Epoch: total number of iterations, one pass over the entire dataset.
model.fit(X_train, y_train, nb_epoch=500, batch_size=120)
```

```python
# Convert the dataframe to a matrix
X_test = X_test.as_matrix()
```

```python
# the prediction results 
classes = model.predict_classes(X_test, batch_size=120)
```

```python
# Print the accuracy of the model
import numpy as np
accuration = np.sum(classes == y_test)/30.0 * 100

print "Test Accuration : " + str(accuration) + '%'
print "Prediction :"
print classes
print "Target :"
print np.asarray(y_test,dtype="int32")
```
